```python
%%writefile app.py
import streamlit as st
import pickle
import pandas as pd # Import pandas for DataFrame operations
import joblib

# Load the movie data and similarity matrix
movies_dict = pickle.load(open('movies.pickle','rb'))
movies = movies_dict['title'].values # Extract movie titles for the selectbox
similarity = pickle.load(open('similarity.pickle','rb')) # Load the similarity matrix
similarity=joblib.load("simlarity.joblib","rb")
# Create a Streamlit web application
# Define the recommendation function
def recommend(movie):
  # Get the index of the selected movie from the movies_dict DataFrame
  movie_index = movies_dict[movies_dict['title'] == movie].index[0]
  # Get the similarity scores for the selected movie
  distances = similarity[movie_index]
  # Sort movies by similarity in descending order and get the top 5 (excluding itself)
  movie_list = sorted(list(enumerate(distances)), reverse=True, key=lambda x:x[1])[1:6]

  recommended_movies = []
  for i in movie_list:
    # Append the title of the recommended movie using its index from movies_dict
    recommended_movies.append(movies_dict.iloc[i[0]].title)
  return recommended_movies

# Streamlit app layout
st.title("Movie Recommendation System")

# Selectbox for choosing a movie
selected_movie_name = st.selectbox(
    "Enter the movie name",
    movies
)

# Button to trigger recommendations
if st.button("Show Recommendations"):
  recommendations = recommend(selected_movie_name)
  st.write("The recommended movies are:")
  for i in recommendations:
    st.write(i)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity

# Initialize CountVectorizer to convert text tags into a matrix of token counts
cv = CountVectorizer(max_features=5000, stop_words='english')
# Fit and transform the 'tags' column to create vectors
vectors = cv.fit_transform(movies_dict['tags']).toarray()

# Compute cosine similarity between movie vectors
similarity = cosine_similarity(vectors)

print("Similarity matrix computed.")
```

```python
import pickle

# Save the similarity matrix to a pickle file
with open('similarity.pickle', 'wb') as f:
    pickle.dump(similarity, f)

print("Similarity matrix saved to similarity.pickle")
```
