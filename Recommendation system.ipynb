```python
import json

def genre_extract(col_value):
  l = []
  if isinstance(col_value, str):
    try:
      items = json.loads(col_value)
    except json.JSONDecodeError:
      return []
  elif isinstance(col_value, list):
    items = col_value
  else:
    return []

  for i in items:
    if isinstance(i, dict) and 'name' in i:
      l.append(i['name'])
    elif isinstance(i, str):
      l.append(i)
  return l
```

```python

data['genres'] = data['genres'].apply(genre_extract)
```

```python
import json

def keywords_extract(col_value):
  s = []
  if isinstance(col_value, str):
    try:
      items = json.loads(col_value)
    except json.JSONDecodeError:
      return []
  elif isinstance(col_value, list):
    items = col_value
  else:
    return []

  for item in items:
    if isinstance(item, dict) and 'name' in item:
      s.append(item['name'])
    elif isinstance(item, str):
      s.append(item)
  return s
```

```python
data['keywords']=data['keywords'].apply(keywords_extract)
```

```python
import json

def cast_extract(col_value):
  l = []
  if isinstance(col_value, str):
    try:
      items = json.loads(col_value)
    except json.JSONDecodeError:
      return []
  elif isinstance(col_value, list):
    items = col_value
  else:
    return []

  for item in items:
    if len(l) < 3:
      if isinstance(item, dict) and 'name' in item:
        l.append(item['name'])
      elif isinstance(item, str):
        l.append(item)
    else:
      break
  return l

data['cast'] = data['cast'].apply(cast_extract)
```

```python
import json

def crew_extract(col_value):
  l = []
  if isinstance(col_value, str):
    try:
      items = json.loads(col_value)
    except json.JSONDecodeError:
      return []
  elif isinstance(col_value, list):
    items = col_value
  else:
    return []

  for item in items:
    if isinstance(item, dict) and item.get('job') == 'Director':
      l.append(item.get('name'))
  return l
```

```python
from nltk.stem.porter  import PorterStemmer
ps=PorterStemmer()
```

```python
data['tags'][0]
```

```python
def stemming(text):
  l=[]
  for i in text.split():
    l.append(ps.stem(i))
  return " ".join(l)
```

```python
data['tags']=data['tags'].apply(stemming)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=5000,stop_words='english')
vectors=cv.fit_transform(data['tags']).toarray()
```

```python
vectors
```

```python
cv.get_feature_names_out()[:200]
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity=cosine_similarity(vectors)
```

```python
similarity
```

```python
data[data['title']=='Avatar'].index[0]
```

```python
def recommend(movie):
  movie_index=data[data['title']==movie].index[0]
  recomendations=similarity[movie_index]
  movie_list=sorted(enumerate(recomendations), reverse=True, key=lambda x:x[1])[1:6]
  for i in movie_list:
    print(data.iloc[i[0]].title)
  return recomendations
```

```python
recommend('Avatar')
```

```python
movie_list=sorted(enumerate(recommend('Avatar')), reverse=True, key=lambda x:x[1])[1:6]
```
